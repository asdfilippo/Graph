```python
import numpy as np
import random
import matplotlib.pyplot as plt
import networkx as nx

# Instanciating a graph
graph = nx.Graph()

# Adding random nodes (maximum of 50)
randomNumber = random.randint(5,51)
listOfNodes = np.random.randint(1, 50, size=(randomNumber))
graph.add_nodes_from(listOfNodes)

# Adding random edges
listOfEdges = []
for i in range(1, randomNumber):
    nodeOrig = random.choice(listOfNodes)
    nodeDest = random.choice(listOfNodes)
    edge = (nodeOrig, nodeDest)
    listOfEdges.append(edge)
graph.add_edges_from(listOfEdges)

# Display graph
nx.draw(graph)
plt.title('Random graph')
plt.show()

print("Number of nodes: ", graph.number_of_nodes())
print("Nodes: ", graph.nodes())
print("\nNumber of edges: ", graph.number_of_edges())
print("Edges: ", graph.edges())
```
